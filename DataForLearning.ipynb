```python
import pandas as pd
import numpy as np
import random as rd

data = pd.read_csv("result.csv", index_col = 0)
data
```

```python
min_x = data.iloc[-1, 0]
max_x = data.iloc[0, 0]
min_y = data.iloc[0, 1]
max_y = data.iloc[-1, 1]
print(min_x, max_x, min_y, max_y)
```

```python
succ = pd.DataFrame({'lat': np.zeros(10000), 'lon': np.zeros(10000), 'success': np.zeros(10000), 'category': np.zeros(10000)})
succ
```

```python
def get_randomX():
    return rd.uniform(min_x, max_x)

def get_randomY():
    return rd.uniform(min_y, max_y)

def get_random_success():
    return rd.uniform(0.1, 0.95)

categories = ['services', 'repair', 'household_products', 'food']

def get_random_category():
    ind = rd.randint(0, len(categories) - 1)
    return categories[ind]
```

```python
succ['lat'] = succ.apply(lambda x: get_randomX(), axis = 1)
succ['lon'] = succ.apply(lambda x: get_randomY(), axis = 1)
succ['success'] = succ.apply(lambda x: get_random_success(), axis = 1)
succ['category'] = succ.apply(lambda x: get_random_category(), axis = 1)
succ
```
