```python
import pandas as pd
import numpy as np

data = pd.read_excel("DataRegion.xlsx")

data.head(5)
```

```python
districts = data["house_district"].unique()
sizes = dict()

for district in districts:
    summ = data[(data["house_district"] == district) & (data["person_exemption_group"])]['sum'].sum()
    sizes[district] = summ

print(sizes)

for index in data.index:
    data.iloc[index, 4] = data.at[index, "sum"] / sizes[data.at[index, "house_district"]]
    
print(data)
```
