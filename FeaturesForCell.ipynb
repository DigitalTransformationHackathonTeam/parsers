```python
import pandas as pd
import numpy as np

social_cat = pd.read_csv('PercentageDataSocialCategories.csv')
print(social_cat)
districts = pd.read_csv('DistrictByCell.csv')
print(districts)
print(social_cat['house_district'].unique())
```

```python
pers_exemp_groups = social_cat['person_exemption_group'].unique()
pers_exemp_groups = np.delete(pers_exemp_groups, 1, 0)

districts = districts.replace(to_replace = 'Хорошёвский', value='Хорошевский', regex=True)
districts = districts.replace(to_replace = 'Савёловский', value='Савеловский', regex=True)

districts
```

```python
ness_dists = districts['District'].unique()

to_drop = list()

for ind, row in social_cat.iterrows():
    if (row['house_district'] not in ness_dists):
        to_drop.append(ind)
social_cat = social_cat.drop(to_drop)
print(social_cat['house_district'].unique())
```

```python
for group in pers_exemp_groups:
    print(group)
    districts[group] = districts.apply(lambda row: social_cat[(social_cat['house_district'] == row['District']) & (social_cat['person_exemption_group'] == group)].sum()['sum'], axis = 1)
    
```
