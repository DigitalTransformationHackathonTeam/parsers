```python
import pandas as pd
import numpy as np

data = pd.read_csv('SocialDataAdded.csv')
population = pd.read_csv('population.csv', index_col = 0)
passability = pd.read_csv('Passability.csv')
metro_distance = pd.read_csv('metro_distance.csv', index_col = 0)

print(data)
print(population.head(5))
print(passability.head(5))
print(metro_distance.head(5))
```

```python
data['cell_traffic'] = passability
data['dist_to_underground'] = list(metro_distance['dist'])
data['population'] = np.random.randint(20, 400, data.shape[0])
data['underground_traffic'] = np.random.randint(200, 3000, data.shape[0])
data
```

```python
data.to_csv(r'C:\Users\1\CellDate.csv', index = False, header=True)
```

```python
data = data.rename(columns = {'Многодетные': 'many_children', 'Пенсионер': 'elders', 'Инвалид': 'disabled'})
data = data.rename(columns = {'f': 'men', '<7': 'children', 'Молодые родители': 'young_parents', 'Беременность': 'pregnant'})
data = data.rename(columns = {'Малоимущие': 'poor', 'Ветеран труда': 'labor_veteran', 'Дети инвалиды': 'disabled_children'})
data = data.drop(columns = {'District', '18-25', '45-60', '60-75', '7-18', '75-90', '90<', '25-45'}, errors = 'ignore')
data
```
